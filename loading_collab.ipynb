```python

from datasets import load_dataset

ds = load_dataset("cfilt/iitb-english-hindi")

print("Splits found:", ds.keys())

def save_split(split_name):
    subset = ds[split_name]
    with open(f"{split_name}.en", "w", encoding="utf-8") as fe, \
         open(f"{split_name}.hi", "w", encoding="utf-8") as fh:
        for x in subset:
            fe.write(x["translation"]["en"].strip() + "\n")
            fh.write(x["translation"]["hi"].strip() + "\n")
    print(f"Saved: {split_name}.en  {split_name}.hi")

for split in ds.keys():
    save_split(split)


```
